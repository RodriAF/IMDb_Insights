# Mergeado simplificado

```python
# Importar
start_time = time.time()
name_basics = pd.read_csv('name.basics.tsv', sep='\t',low_memory=False)
title_akas = pd.read_csv('title.akas.tsv', sep='\t',low_memory=False)
title_basics = pd.read_csv('title.basics.tsv', sep='\t',low_memory=False)
title_crew = pd.read_csv('title.crew.tsv', sep='\t',low_memory=False)
title_episode = pd.read_csv('title.episode.tsv', sep='\t',low_memory=False)
title_principals = pd.read_csv('title.principals.tsv', sep='\t',low_memory=False)
title_ratings = pd.read_csv('title.ratings.tsv', sep='\t',low_memory=False)
print("--- %s seconds ---" % (time.time() - start_time))
```

```python
# Quitar tvEpisode y peliculas para adultos
start_time = time.time()
new_title_basics = title_basics[(title_basics.titleType != "tvEpisode")]
new_title_basics = new_title_basics[(new_title_basics.isAdult == "0")]
print("--- %s seconds ---" % (time.time() - start_time))
```

```python
# Cambio a tconst de titleld en akas
start_time = time.time()
title_akas.columns = ['tconst', 'ordering', 'title', 'region', 'language', 'types',
       'attributes', 'isOriginalTitle']
print("--- %s seconds ---" % (time.time() - start_time))
```

```python
# Inner merges on tconst and nconst
start_time = time.time()
title = pd.merge(new_title_basics,title_crew,on='tconst',how='inner')
title = pd.merge(title,title_ratings,on='tconst',how='inner')
#title = pd.merge(title,title_principals,on='tconst',how='inner')
#title = pd.merge(title,name_basics,on='nconst',how='inner')
title = pd.merge(title,title_akas,on='tconst',how='inner')
print("--- %s seconds ---" % (time.time() - start_time))
```

```python
# Export as tsv or csv
start_time = time.time()
# title.to_csv('title_final_inner.csv',sep=",")
title.to_csv('rodri_title_final_inner.tsv',sep="\t")
print("--- %s seconds ---" % (time.time() - start_time))
```
